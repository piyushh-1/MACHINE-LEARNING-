# Q1 What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

ANS1 Anomaly Detection is a technique in Machine Learning used to identify data points, patterns, or events that deviate significantly from normal behavior. These unusual patterns are called anomalies, outliers, or exceptions.

It is widely used in areas like fraud detection, network security, medical diagnosis, and fault detection.


# Types of Anomalies
#1. Point Anomalies (Global Anomalies)

A point anomaly is a single data point that is significantly different from the rest of the data.

# Example:

In a dataset of daily temperatures (25°C–35°C), a value of 60°C is a clear anomaly.
In banking, if a user usually spends ₹500–₹2000 but suddenly makes a transaction of ₹5 lakh, it's a point anomaly.

# Key Idea:
One data point is abnormal compared to the entire dataset.


#2. Contextual Anomalies (Conditional Anomalies)

A contextual anomaly depends on the context (such as time, location, or season). A data point may be normal in one situation but abnormal in another.

# Example:

A temperature of 15°C:
Normal in winter ❄️
Abnormal in summer ☀️
High website traffic at 2 PM is normal, but the same traffic at 3 AM might be suspicious.

# Key Idea:
Anomaly depends on contextual information (time, season, etc.).


#3. Collective Anomalies

A collective anomaly occurs when a group of data points together behaves abnormally, even if individual points seem normal.

# Example:

A sudden spike in network traffic over a period may indicate a cyberattack.
In ECG data, a sequence of normal-looking heartbeats forming an unusual pattern could signal a health issue.

# Key Idea:
A pattern or group of points is anomalous, not individual values.

#Summary Table


| Type               | Description                       | Example                |
| ------------------ | --------------------------------- | ---------------------- |
| Point Anomaly      | Single unusual data point         | ₹5 lakh transaction    |
| Contextual Anomaly | Depends on context (time/season)  | 15°C in summer         |
| Collective Anomaly | Group of points behave abnormally | Network attack pattern |


# Q2  Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases



ANS2 Here’s a clear comparison of three popular anomaly detection methods—Isolation Forest, DBSCAN, and Local Outlier Factor (LOF)—focusing on how they work and when to use them.

#1. Isolation Forest
# Approach
.Based on the idea that anomalies are easier to isolate than normal points.
.Builds multiple random decision trees.
.Points that require fewer splits to isolate are likely anomalies.
# Intuition

Anomalies are rare and different → they get separated quickly.

# Suitable Use Cases
.High-dimensional datasets
.Large-scale data
.Fraud detection (e.g., unusual transactions)
.When you don’t know data distribution
# Pros
.Fast and scalable
.Works well with high dimensions
.No need for distance calculations



# Cons
.Less effective when anomalies are not clearly separable
.Doesn’t consider local density

#2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
# Approach
.Groups points based on density.
.Points in dense regions → clusters
.Points in sparse regions → anomalies (noise)
# Intuition

Normal data forms dense clusters, anomalies lie alone.

# Suitable Use Cases
.Spatial data (e.g., GPS, geolocation)
.Clustering + anomaly detection together
.When clusters have arbitrary shapes
# Pros
.No need to specify number of clusters
.Detects noise naturally
.Works well with irregular cluster shapes

# Cons
.Sensitive to parameters (eps, min_samples)
.Struggles with varying densities
.Not ideal for high-dimensional data

#3. Local Outlier Factor (LOF)
# Approach
.Measures local density deviation of a data point compared to its neighbors.
.If a point has much lower density than neighbors → anomaly.
# Intuition

Anomalies are points that are less dense than their surroundings.

# Suitable Use Cases
.Detecting local anomalies
.Data with varying density regions
.Intrusion detection, fraud detection
# Pros
.Captures local structure well
.Better than global methods for complex datasets

# Cons
.Computationally expensive
.Sensitive to number of neighbors (k)
.Not great for very large datasets

#Side-by-Side Comparison

| Feature               | Isolation Forest 🌲    | DBSCAN 📍                 | LOF 🔎                   |
| --------------------- | ---------------------- | ------------------------- | ------------------------ |
| Approach              | Tree-based isolation   | Density-based clustering  | Local density comparison |
| Detects anomalies as  | Easily isolated points | Noise points              | Low-density points       |
| Works on              | Global anomalies       | Spatial/density anomalies | Local anomalies          |
| Scalability           | High                   | Medium                    | Low                      |
| High-dimensional data | ✔️ Good                | ❌ Poor                    | ❌ Limited                |
| Parameters            | Few                    | Sensitive (eps, minPts)   | Sensitive (k neighbors)  |
| Cluster detection     | ❌ No                   | ✔️ Yes                    | ❌ No                     |


# Q3 What are the key components of a Time Series? Explain each with one example.


ANS3 A Time Series is a sequence of data points collected or recorded at regular time intervals (daily, monthly, yearly, etc.).
Understanding its components helps in analysis, forecasting, and pattern recognition.

#Key Components of a Time Series
#1. Trend (T)
# What it is:

The long-term movement or direction in the data over time (upward, downward, or stable).

# Example:
.A company’s sales increasing every year due to business growth.
.Population growth over decades.

# Key Idea:
Shows the overall direction of data.

#2. Seasonality (S)
# What it is:

Regular and repeating patterns occurring at fixed intervals (daily, monthly, yearly).

# Example:
.Ice cream sales increasing every summer 🍦
.Higher shopping during festivals like Diwali

# Key Idea:
Patterns repeat over a known, fixed period.

#3. Cyclical Component (C)
# What it is:

Fluctuations that occur over longer, irregular periods, often linked to economic or business cycles.

# Example:
.Rise and fall of stock markets during economic booms and recessions
.Business profits increasing during growth phases and decreasing during downturns

# Key Idea:
Not fixed like seasonality, but follows economic cycles.

#4. Irregular (Random/Noise) Component (I)
# What it is:

Unpredictable, random variations that cannot be explained by trend, seasonality, or cycles.

# Example:
.Sudden drop in sales due to a natural disaster
.Unexpected spike in demand due to viral social media trends

# Key Idea:
Random, uncontrollable fluctuations.


#Summary Formula

A time series is often represented as:

# Additive Model:

Y=T+S+C+I



















#Multiplicative Model:
Y=T×S×C×I

# What is Stationarity in Time Series?

A time series is called stationary if its statistical properties remain constant over time.

# Key properties:
.Mean → constant
.Variance → constant
.Autocorrelation → depends only on lag, not time

# In simple words:
The data doesn’t show trends, seasonal effects, or changing variability over time.

#Why is Stationarity Important?

Many models in Time Series Analysis—like ARIMA—assume the data is stationary.
Non-stationary data can lead to incorrect predictions and misleading patterns.

#How to Test Stationarity?
#1. Visual Inspection 👀
.Plot the time series
.Check for:
.Trend 📈
.Seasonality 🔁
.Changing variance

# Example:

.Upward trend → non-stationary
.Constant fluctuations around mean → stationary

.Most commonly used test
# Hypotheses:
.Null Hypothesis (H₀): Series is non-stationary
.Alternative (H₁): Series is stationary
# Decision Rule:
.If p-value < 0.05 → Reject H₀ → Stationary
.If p-value ≥ 0.05 → Non-stationary

#Other Tests:
.KPSS Test
.Phillips-Perron Test

# Q7 Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Reproducibility
np.random.seed(42)

# --- Create synthetic dataset (distance vs fare) ---
n_normal = 500

# Trip distance (km)
distance = np.random.gamma(shape=2.0, scale=2.0, size=n_normal)

# Fare ≈ base + rate * distance + noise
base_fare = 2.5
rate_per_km = 1.8
noise = np.random.normal(0, 2.0, size=n_normal)
fare = base_fare + rate_per_km * distance + noise

# Add anomalies
n_anom = 25
distance_anom = np.random.uniform(0.1, 20, size=n_anom)
fare_anom = np.concatenate([
    np.random.uniform(60, 120, size=n_anom // 2),   # very high fares
    np.random.uniform(-5, 2, size=n_anom - n_anom // 2)  # unrealistic fares
])

# Combine dataset
distance_all = np.concatenate([distance, distance_anom])
fare_all = np.concatenate([fare, fare_anom])

df = pd.DataFrame({
    "distance_km": distance_all,
    "fare": fare_all
})

# --- Apply Isolation Forest ---
model = IsolationForest(contamination=0.05, random_state=42)
df["anomaly"] = model.fit_predict(df[["distance_km", "fare"]])

# Separate normal and anomalies
normal = df[df["anomaly"] == 1]
anomalies = df[df["anomaly"] == -1]

# --- Plot ---
plt.figure()
plt.scatter(normal["distance_km"], normal["fare"], label="Normal")
plt.scatter(anomalies["distance_km"], anomalies["fare"], label="Anomaly")
plt.title("Isolation Forest: Taxi Fare Anomaly Detection")
plt.xlabel("Distance (km)")
plt.ylabel("Fare")
plt.legend()
plt.show()

# Show detected anomalies
print(anomalies.head())
```

# Q9 Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.
(Include your Python code and output in the code box below.)



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs

# --- 1. Generate a synthetic numerical dataset ---
# Generate data with two dense clusters (inliers)
n_samples = 300
outliers_fraction = 0.05
n_outliers = int(outliers_fraction * n_samples)
n_inliers = n_samples - n_outliers

blobs_params = dict(random_state=42, n_samples=n_inliers, n_features=2)
X, _ = make_blobs(centers=[[0, 0], [0, 0]], cluster_std=0.5, **blobs_params)

# Add some random uniform noise (outliers)
rng = np.random.RandomState(42)
X = np.concatenate([X, rng.uniform(low=-6, high=6, size=(n_outliers, 2))], axis=0)

# --- 2. Apply the Local Outlier Factor (LOF) algorithm ---
# n_neighbors determines the local neighborhood size.
# contamination is the expected proportion of outliers in the dataset.
clf = LocalOutlierFactor(n_neighbors=20, contamination=outliers_fraction)

# Use fit_predict to compute the predicted labels of the training samples
# y_pred will be 1 for inliers and -1 for outliers
y_pred = clf.fit_predict(X)

# Get the negative outlier factors (LOF scores)
X_scores = clf.negative_outlier_factor_
# The actual LOF score is the negative of X_scores.
# Scores close to 1 are normal, scores significantly higher than 1 (meaning X_scores are significantly lower than -1) are outliers.

# --- 3. Visualize the results using matplotlib ---
plt.figure(figsize=(12, 6))

# Scatter plot of data points, colored by their prediction
# Blue for inliers (y_pred == 1), Red for outliers (y_pred == -1)
colors = np.array(['#377eb8', '#ff7f00']) # Blue, Orange (we'll use 'red' for explicit outliers in the legend)
plt.scatter(X[:, 0], X[:, 1], color=colors[(y_pred + 1) // 2], s=10, label='Data points') # Using a simple color mapping with a better aesthetic

# Plot circles with radius proportional to the outlier scores to visually emphasize abnormality
# The smaller the negative outlier factor (more negative), the higher the degree of abnormality, so we use the inverse for radius scaling
radius = (X_scores.max() - X_scores) / (X_scores.max() - X_scores.min())
scatter = plt.scatter(X[:, 0], X[:, 1], s=1000 * radius, edgecolors='r', facecolors='none', alpha=0.5, label='Outlier scores visualization')

plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
# Create a custom legend entry for outliers
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='#377eb8', label='Inliers (+1 prediction)'),
    Patch(facecolor='#ff7f00', label='Outliers (-1 prediction)'),
    Patch(edgecolor='r', facecolor='none', linewidth=1, label='Higher LOF score (larger circle)')
]
plt.legend(handles=legend_elements, loc="upper left")
plt.axis('tight')
plt.show()

# --- 4. Print identified outliers ---
outliers = X[y_pred == -1]
print(f"Number of anomalies detected: {len(outliers)}")
print("Coordinates of some detected anomalies (first 5):")
print(outliers[:5])
```

# Q10 You are working as a data scientist for a power grid monitoring company.Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage.

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# Simulated time series (15-min intervals)
rng = pd.date_range("2024-01-01", periods=500, freq="15T")
np.random.seed(42)

# Simulated energy usage with seasonality
energy = 50 + 10*np.sin(np.linspace(0, 20, 500)) + np.random.normal(0, 2, 500)

# Add anomalies
energy[100] = 100  # spike
energy[300] = 10   # drop

df = pd.DataFrame({"energy": energy}, index=rng)

# --- Anomaly Detection ---
iso = IsolationForest(contamination=0.02)
df["anomaly"] = iso.fit_predict(df[["energy"]])

# --- Forecasting ---
model = SARIMAX(df["energy"], order=(1,1,1), seasonal_order=(1,1,1,96))
results = model.fit(disp=False)

forecast = results.get_forecast(steps=50)
forecast_mean = forecast.predicted_mean

# --- Plot ---
plt.figure()
plt.plot(df["energy"], label="Actual")
plt.plot(forecast_mean, label="Forecast")
plt.scatter(df[df["anomaly"]==-1].index,
            df[df["anomaly"]==-1]["energy"],
            label="Anomaly")
plt.legend()
plt.show()

# Output sample
print(df[df["anomaly"]==-1].head())
```
